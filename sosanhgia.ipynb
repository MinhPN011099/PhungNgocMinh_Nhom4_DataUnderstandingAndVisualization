```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('sosanhgia.csv')
```

```python
data
```

```python
data.fillna(np.int(0), inplace = True)
```

```python
data['name'][data['category/1'] == 'Laptop']
```

```python
data.dtypes
```

```python
data.shape
```

Convert price columns from string to int

```python
mod_price = []
for price in data['priority_price']:
    temp = str(price).replace('.', '')
    mod_price.append(int(temp))
data['mod_price'] = pd.Series(mod_price)
```

Tìm chiếc Laptop có giá thấp nhất

```python
laptop = data[['name', 'mod_price']][data['category/1'] == 'Laptop']
```

```python
print("The cheapest laptop is: {} - {}".format(laptop['name'][laptop['mod_price'].idxmin()], laptop['mod_price'].min()))
      
      
```

```python
x = pd.Series(laptop['mod_price'], name = 'Distribution of Prices')
dist_price = sns.distplot(x, color = 'b')
```

Phân tích Tiki

```python
tiki = data[['category/0', 'name', 'mod_price']][data['priority_store'] == 'Tiki']
tiki
```

```python
x = tiki['category/0'].value_counts()
category = tiki['category/0'].unique()
avg_price = []
for cat in category:
    avg_price.append(data['mod_price'][data['category/0'] == cat].mean())
avg_price
```

```python
avg_price_df = pd.DataFrame(avg_price, index = category, columns = ['Avg Price'])
avg_price_df
```
