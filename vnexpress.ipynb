```python
df = data.drop(['tags/5', 'tags/6', 'tags/7', 'tags/8', 'tags/9', 'tags/10', 'tags/11', 'tags/12', 'tags/13', 'tags/14', 'tags/15'], axis = 1)
```

Đưa ra các tin tức liên quan đến Covid-19

```python
cond1 = df['tags/0'] == 'covid-19'
cond2 = df['tags/1'] == 'covid-19'
cond3 = df['tags/2'] == 'covid-19'
cond4 = df['tags/3'] == 'covid-19'
cond5 = df['tags/4'] == 'covid-19'
covid_19 = df[['link', 'title', 'description', 'pub_date']][cond1 | cond2 | cond3 | cond4 | cond5]
covid_19
```

Đưa ra các tin tức về Covid-19 trong tháng 8/2020

```python
cond = []
for i in covid_19.index:
    try:
        match = re.search('.+/8/.+', covid_19['pub_date'][i])
        if type(match) == re.Match:
            cond.append(True)
        else:
            cond.append(False)
    except:
        cond.append(False)
```
